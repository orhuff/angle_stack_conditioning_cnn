```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import tensorflow as tf
import matplotlib.patches as patches
import segyio
```

```python
far_orig_file = 'far_cube.sgy'

far_cnn_file = 'far_cube_cnn.sgy'
```

```python
far_orig = segyio.tools.cube(far_orig_file)

far_cnn = segyio.tools.cube(far_cnn_file)
```

```python
far_orig.shape
# IL, XL, time
```

```python
fig,ax = plt.subplots(figsize=(7,7))
yticks = ticker.FuncFormatter(lambda y, pos: '{0:g}'.format(y*4))
xticks = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(34895+x))
c = 0.16
ax.imshow(far_orig[:, 1, :].T/np.max(far_orig[:, 1, :]), aspect='auto', cmap='Greys', vmin=-c, vmax=c)
ax.xaxis.set_major_formatter(xticks);
ax.yaxis.set_major_formatter(yticks);
ax.set_xlabel('IL Number')
ax.set_ylabel('Time (ms)')

#fig.savefig('far_orig_1.jpg', dpi=1000, bbox_inches='tight')
```

```python
fig,ax = plt.subplots(figsize=(7,7))


ax.imshow(far_cnn[:, 1, :].T/np.max(far_cnn[:, 1, :]), aspect='auto', cmap='Greys', vmin=-c, vmax=c)
ax.xaxis.set_major_formatter(xticks);
ax.yaxis.set_major_formatter(yticks);
ax.set_xlabel('IL Number')
ax.set_ylabel('Time (ms)')

#fig.savefig('far_cnn_1.jpg', dpi=1000, bbox_inches='tight')
```

```python
fig,ax = plt.subplots(figsize=(7,7))
start = 250
end = 900
yticks_zoom = ticker.FuncFormatter(lambda y, pos: '{0:g}'.format((start+y)*4))
xticks_zoom = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(34895+x))

ax.imshow(far_orig[:, 1, start:end].T/np.max(far_orig[:, 1, :]), aspect='auto', cmap='Greys', vmin=-c, vmax=c)

ax.xaxis.set_major_formatter(xticks_zoom);
ax.yaxis.set_major_formatter(yticks_zoom);
ax.set_xlabel('IL Number')
ax.set_ylabel('Time (ms)')
#fig.savefig('far_orig_1_zoom.jpg', dpi=1000, bbox_inches='tight')
```
